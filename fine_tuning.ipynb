```python
import torch

from transformers import AutoModelForQuestionAnswering
from torch.utils.data import DataLoader, random_split
from data.dataset import ExtractMatchDataset
from run_train import run_train
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
train_data, val_data = random_split(
    ExtractMatchDataset(),
    [0.8, 0.2],
    generator=torch.Generator().manual_seed(42)
)
```

```python
train_dataset = DataLoader(train_data, batch_size=32, shuffle=True, num_workers=4 or 8)
val_dataset = DataLoader(val_data, batch_size=32, shuffle=True, num_workers=4 or 8)
```

```python
model = AutoModelForQuestionAnswering.from_pretrained("MilyaShams/rubert-russian-qa-sberquad").to(device)
```

```python
history = run_train(
    model,
    train_dataset,
    val_dataset,
    20
)
```
