```python
import pandas as pd
import matplotlib.pyplot as plt

from data.dataset import ExtractMatchDataset
```

```python
dataset = ExtractMatchDataset()
```

```python
analyzis_blocks = []

for data in dataset.get_data():
    cls = 1 if data[1] == "обеспечение исполнения контракта" else 0
    block = [cls, data[2][1] - data[2][0]]
    analyzis_blocks.append(block)
```

```python
df = pd.DataFrame(analyzis_blocks, columns=["class", "symbols"])
```

```python
df.head()
```

```python
plt.hist(df["symbols"], bins=20, edgecolor="black")
plt.grid()
```

```python
classes = [1, 0]
sizes = [df[df["class"] == 1].shape[0], df[df["class"] == 0].shape[0]]

plt.pie(
    sizes, 
    labels=classes, 
    autopct="%1.1f%%",
    explode=[0.01, 0.1],
    shadow=True
    )

labels = [f"1 - {sizes[0]/df.shape[0]*100:.1f}%", f"0 - {sizes[1]/df.shape[0]*100:.1f}%"]
plt.legend(labels=labels)
plt.title("Distribution by tags")
plt.show()
```
